# Data Visualization Overview

## Matplotlib Basics
---

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
x = [0, 1, 2]
y = [100, 200, 300]
```

```python
plt.plot(x,y)
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                       'price':[100,120,190,200,230,310,330,305]})
```

```python
housing
```

```python
plt.scatter(housing['rooms'],housing['price'])
plt.title('Housing Price')
plt.xlabel('Rooms')
plt.ylabel('Price')
```

```python
x = [0, 1, 2]
y = [100, 200, 300]
```

```python
plt.plot(x,y)
plt.xlim(0,5)
plt.ylim(100,500)
plt.plot(x,y,color='red', marker = 'o', linestyle = '--')
```

## Seaborn Basics
----

```python
import seaborn as sns
```

```python
df = pd.read_csv('DATA/heart.csv')
```

```python
df.head()
```

```python
plt.figure(figsize = (6,4))
sns.histplot(df['age'], bins = 40)
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='target',data=df)
```

```python
sns.boxplot(x='sex',y='age',data=df)
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df)
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df,hue='sex')
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df,hue='sex',palette='Dark2', size = 'age')
```

```python
iris = pd.read_csv('DATA/iris.csv')
```

```python
iris.head()
```

```python
sns.countplot(x = 'species', data = iris)
```

```python
sns.pairplot(iris, hue = 'species')
```

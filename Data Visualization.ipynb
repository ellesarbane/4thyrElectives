## Data Visualization

## Matplotlib
-----

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
#% matplotlib inline - for older version
```

```python
x = [0, 1, 2]
y = [100, 200, 300]
```

```python
plt.plot(x,y) # returns an object or add semi colon to remove 
plt.show() # act as print
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3,],
                       'price':[100,120,190,200,230,310,330,305]})
housing
```

```python
plt.scatter(housing['rooms'],housing['price'])
plt.title('Housing Price')
plt.xlabel('Rooms')
plt.ylabel('Price')
```

```python
plt.plot(x,y)
plt.xlim(0,5)
plt.ylim(100,500)
plt.plot(x,y,color='red', marker = 'o', linestyle = '--') #change color, line,
```

## Seaborn
-----

```python
import seaborn as sns
```

```python
df = pd.read_csv('Data/heart.csv')
df
```

```python
df.head() #first 5, can be a specific number
df.tail(10) #last 5
```

```python
#plt.figure(figsize = (12,8)) #change size
sns.displot(df['age'], bins = 40) #distribution, more bins = thin blocks
```

```python
sns.countplot(x = 'sex', data = df)
```

```python
sns.countplot(x = 'target', data = df)
```

```python
sns.boxplot(x = 'sex', y = 'age', data = df) #quartile range, used by future engineer
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df, hue='sex') #
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df, hue='sex', palette='Dark2')
```

```python
sns.scatterplot(x='chol',y='trestbps',data=df, hue='sex', palette='Dark2', size = 'age')
```
